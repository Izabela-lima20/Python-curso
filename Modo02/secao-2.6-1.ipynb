```python
## Desenvolver um programa para solicitar o valor de um temperatura em grau Celsos e convertela para grauas Fahrenheit
# em graus Celsius e converte-lá para graus Fahrenheit
# a formula de conversão é:
# F = C * 9/5 + 32

celsius = float(input("Diigte a temperatura em celsius:")) # aqui estou convertento celsis para float, e pedindo para o usuario digitar a temperatura
fahrenheit = celsius * 9/5 + 32 # aqui e a formula
print("a temperatura em Fahrenheit é: ", fahrenheit, "F", sep="") 
```

## 2.6.11   LAB   Operadores e Expressões – 2


```python
#Resolução

hour = int(input("Hora de início (horas): ")) #A primeira linha solicita ao usuário que insira a hora de início do evento (em horas).
mins = int(input("Hora de início (minutos): ")) #  A segunda linha pede ao usuário a quantidade de minutos da hora de início.
dura = int(input("Duração do evento (minutos): ")) # A terceira linha solicita a duração do evento em minutos.

tempo_total = hour * 60 + mins + dura #Aqui, o código calcula o tempo total do evento em minutos.
hora_final = tempo_total // 60 % 24 # Essa linha calcula a hora final (em formato de 24 horas).
minitos_final = tempo_total % 60 #  Aqui, os minutos finais são obtidos ao calcular o resto da divisão do tempo total por 60 (o que dá os minutos dentro da hora).

print(hora_final, minitos_final , sep=":") # Por fim, o código exibe o horário final no formato hora:minuto, usando sep=":" para separar a hora e os minutos com dois pontos.

#hour * 60 converte a hora de início para minutos.

#mins são os minutos da hora de início.

#dura é a duração do evento em minutos. O resultado é o total de minutos desde meia-noite (00:00) até o término do evento.

```
