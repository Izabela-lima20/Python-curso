```python
# exemplo para testar maior valor 
# Usuario vai digitar os valores ate o valor ser  -1

maior = -999_999_999_999
valor = int(input("Digite um valor (-1 = fim):"))
while valor != -1:
    if valor>maior:
        maior = valor

    valor = int(input("Digite um valor (-1 = fim):"))
if maior > -999_999_999_999:
   print("o maior valor digitado foi", maior)

else:
    print("não houve digitação de valores!")
```

```python
# solicitar algo para o usuario a digitação de valor, se -1 encerrar
#contar quantos pares e quantos impares foram difitados 

par, impar = 0,0 # estou atribuindo o par e impar para 0
num = int(input("Digite uma numero (-1 = fim):"))
while num != -1:
    if num % 2 == 0: # par
        par +=1 # par = par + 1
    else:
        impar += 1 # impar = impar + 1 
    num = int(input("Digite uma numero (-1 = fim):"))
print("Pares:", par, "impares:", impar)
```

### As instruções break e continue
-break - sai do loop imediatamente e termina incondicionalmente a operação do loop; o programa começa a executar a instrução mais próxima após o corpo do loop;
-continue - se comporta como se o programa tivesse chegado ao fim do corpo; o próximo turno é iniciado e a expressão de condição é testada imediatamente.

```python
contador = 0
while contador < 10: 
    print(contador)
    contador += 1
    if contador == 3:
        break
print("Fim do programa")
```

```python
contador = 0
while contador < 10: 
    contador += 1
    if contador == 3:
         continue
    print(contador)
print("Fim do programa")
```
