```python
# exemplo para testar maior valor 
# Usuario vai digitar os valores ate o valor ser  -1

maior = -999_999_999_999
valor = int(input("Digite um valor (-1 = fim):"))
while valor != -1:
    if valor>maior:
        maior = valor

    valor = int(input("Digite um valor (-1 = fim):"))
if maior > -999_999_999_999:
   print("o maior valor digitado foi", maior)

else:
    print("não houve digitação de valores!")
```

```python
# solicitar algo para o usuario a digitação de valor, se -1 encerrar
#contar quantos pares e quantos impares foram difitados 

par, impar = 0,0 # estou atribuindo o par e impar para 0
num = int(input("Digite uma numero (-1 = fim):"))
while num != -1:
    if num % 2 == 0: # par
        par +=1 # par = par + 1
    else:
        impar += 1 # impar = impar + 1 
    num = int(input("Digite uma numero (-1 = fim):"))
print("Pares:", par, "impares:", impar)
```

### As instruções break e continue
-break - sai do loop imediatamente e termina incondicionalmente a operação do loop; o programa começa a executar a instrução mais próxima após o corpo do loop;
-continue - se comporta como se o programa tivesse chegado ao fim do corpo; o próximo turno é iniciado e a expressão de condição é testada imediatamente.

```python
contador = 0
while contador < 10: 
    print(contador)
    contador += 1
    if contador == 3:
        break
print("Fim do programa")
```

```python
contador = 0
while contador < 10: 
    contador += 1
    if contador == 3:
         continue
    print(contador)
print("Fim do programa")
```

### Laboratorio 3.2.14


```python
# resolução

blocks = int(input("Insira o número de blocos:")) #Aqui, o programa pede para o usuário inserir o número total de blocos que ele tem disponível para construir a pirâmide. Esse valor é convertido para um número inteiro e armazenado na variável

altura = 0  #Duas variáveis são inicializadas:
            #altura: vai armazenar a altura da pirâmide.
            #total_blocos: vai armazenar o número total de blocos usados até o momento.
total_blocos = 0


while total_blocos + altura + 1 <= blocks: #O loop while vai continuar enquanto a soma do total de blocos usados mais o número de blocos necessários para o próximo nível da pirâmide (altura + 1) for menor ou igual ao número total de blocos disponíveis (blocks). Ou seja, ele verifica se ainda é possível adicionar um novo nível à pirâmide.

    altura += 1 # A altura da pirâmide é aumentada em 1 (representando que um novo nível foi adicionado).
    total_blocos += altura #O número de blocos usados até aquele ponto (total_blocos) é aumentado pelo valor de altura (porque o primeiro nível usa 1 bloco, o segundo usa 2 blocos, o terceiro usa 3 blocos, e assim por diante).
 
print("A altura da pirâmide:", altura)

```

```python
# fazer um programa que solicite um numero e retorne a raiz aproximada do numero digitado. Deve usar laço for, if e break
# Resolução

num = int(input("Digite o valor:"))
raiz = 0
for i in range(num):
        raiz = i
        if (raiz + 1) * (raiz + 1) > num:
                 break
print("Raiz aproximada é:" , raiz)
```
