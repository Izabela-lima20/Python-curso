```python
## Fatiamento
lista = [1,2,3,4,5]
print(lista[1:3])# Esse código cria uma fatia da lista, começando do índice 1 até o índice 3 (mas não inclui o valor do índice 3). Em outras palavras, pega os elementos do índice 1 até o índice 2.Resultado: [2, 3]
print(lista[:3])# Aqui, a fatia começa no início da lista e vai até o índice 3 (novamente, não inclui o valor do índice 3). Ou seja, pega os elementos do índice 0 até o índice 2.Resultado: [1, 2, 3]
print(lista[1:]) # Nesse caso, a fatia começa no índice 1 e vai até o final da lista. Como não há valor final especificado, ela pega todos os elementos a partir do índice 1 até o final. Resultado: [2, 3, 4, 5]
print(lista[:])#A fatia aqui não tem índice inicial nem final, o que significa que pega todos os elementos da lista, do começo ao fim. Resultado: [1, 2, 3, 4, 5]
print(lista[1:-2])# A fatia começa no índice 1 e vai até o penúltimo índice, já que o valor -2 refere-se ao segundo item a partir do final da lista. Ou seja, exclui o último e o penúltimo elemento. Resultado: [2, 3]
```

```python
lista1 = [1,2,3,4]# Criando uma lista chamada lista1
lista2 = lista1[:]# Criando uma cópia da lista1 usando fatiamento
lista2.append(5) # Adicionando o valor 5 à lista2
print(lista1)
print(lista2)
#O que aconteceu aqui
#lista2 = lista1[:] cria uma cópia independente de lista1. As alterações feitas em lista2 (como adicionar o número 5) não afetam lista1, porque elas são listas separadas após a cópia. 
# o fatiamento cria uma copia 
# cada elemento da lista 
```

```python
lista1 = [1,2,3,4]# Criando uma lista chamada lista1
lista2 = lista1[:]# Criando uma cópia da lista1 usando fatiamento
lista2.append(5) # Adicionando o valor 5 à lista2
print(lista1,id(lista1))
print(lista2,id(lista2))
del lista1[1:3] # Deleta os elementos no índice 1 até 2 de lista1
print(lista1)
print(lista2)
```

### operadores in e not in

```python
lista = [7,8,3,2,5]# Criando uma lista chamada lista com os elementos [7, 8, 3, 2, 5]
print(5 in lista)  # Verifica se o valor 5 está na lista
print(5 not in lista) # Verifica se o valor 5 NÃO está na lista
print(9 in lista)  # Verifica se o valor 9 está na lista
```

```python
#fatiamento em string
texto = "abcdefghijklmnopqrstuvwxyz"
print(texto[2:-2])
```

### 3.6.6   LAB   Operações com listas ‒ básico



```python
#resolução

minha_lista = [1, 2, 4, 4, 1, 4, 2, 6, 2, 9]

lista_unica= [] # Lista para armazenar elementos únicos, sem repetições /  é uma nova lista vazia que será usada para armazenar os números únicos.

for num in minha_lista:# Para cada número em minha_lista

    if num not in lista_unica:# Se o número não estiver na lista lista_unica

        lista_unica.append(num)  # Adiciona o número a lista_unica
        
        minha_lista = lista_unica# A lista original agora é igual à lista única sem repetições

print("A lista com os elementos são")
print(minha_lista)
```

```python
# OUTRA FORMA

my_list = [1, 2, 4, 4, 1, 4, 2, 6, 2, 9]

apoi=[]
for el in my_list:
    if el not in apoi:
        apoi.append(el)
        my_list = apoi
print("A lista com os elementos exclusivos aqui")
print(my_list)
```
