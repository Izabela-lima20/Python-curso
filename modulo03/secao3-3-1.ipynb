### operador bit-a-bit 
##### operador de conjunção ( & ) ou and bit-a-bit 

```python
2 & 3 = 2
2 10 &
3 11 & 
-----
2 10
```

```python
a = 5
b = 7
print(a,f"{a:04b}") #O print mostra o valor de a seguido de sua representação binária com pelo menos 4 bits (:04b). O b significa binário e o 04 garante que a saída tenha 4 dígitos, preenchendo com zeros à esquerda, se necessário./ 5 em binário é 101. Com a formatação 04b, ele é exibido como 0101.
print(b,f"{b:04b}", "&") #Aqui, estamos fazendo algo similar com b, que é 7: / 7 em binário é 111. Com a formatação 04b, ele será exibido como 0111.
print("-"*7) # isso e  -----
print(a&b,f" {(a&b):04b}") #Aqui, a operação AND bit a bit é realizada entre a e b. O operador & compara os bits de a e b na mesma posição: a = 5 (0101 em binário) b = 7 (0111 em binário)
```

### operador de disjunção (|) ou or bit-a-bit
## 1 byte = 8 bits
### "a" 
5 0101
7 0111
-------
7 0111
### se tiver 1 vai ser 1 (olha eu queo ligar tal coisa independente se já estiver ligado)

```python
a = 5
b = 7
print(a,f"{a:04b}")
print(b,f"{b:04b}", "&")
print("-"*7)
print(a|b,f"{(a|b):04b}")
```

### operador xor (^) ou or exclusivo bit-a-bit
5 0101
7 0111
-------
2 0010

```python
a = 5
b = 7
print(a,f"{a:04b}")
print(b,f"{b:04b}", "&")
print("-"*7)
print(a^b,f"{(a^b):04b}")
```

### complemento (~)
~5 = -6
### sem um numero a mais negativo 
## exemplo ~4 = -5
### ele aumenta um e inverte o sinal


```python
print(~5)
print(~4)
```

### deslocameto de bit para a esquerda (<<)
1 << 2 resultado em 100

```python
print(1 << 2, f"{(1<<2):04b}")
# ele faz deslocamento para a esquerda aqui deslocamento de 2 casas 
# exemplo 5<<2 ou sera 5 x 2 elevado a 2
 #5 * 2 (esse 2 e elevaod a 2) = 4 
```

```python
print(1 << 2, f"{(5<<2):04b}")
```

### deslocameto de bit para a direita (>>)
5 >> 2 resultado em 1 
## ele pedi os digitos 
0101 eu elimino os dois utlimos bits (01) 

```python
print(5 >> 2, f"{(5>>2):04b}")
```
